```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, log_loss, roc_auc_score, roc_curve, accuracy_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df
```

```python
df['category'] = wine.target
df
```

```python
df.info
```

```python
df['category'].value_counts()
```

```python
df
```

```python
df.isnull().sum()
```

```python
df
```

```python
x = df.drop('category', axis=1)
y = df['category']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state = 7, stratify=y)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
df['category'].value_counts()
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
model.classes_
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
confusion_matrix(y_test, y_pred)
```

```python
model.score(x_test,y_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import multilabel_confusion_matrix

multilabel_confusion_matrix(y_test, y_pred) # To get true_+ve_rate, ture_-ve_rate
```

```python
y_pred_prob = model.predict_proba(x_test)
y_pred_prob
```
